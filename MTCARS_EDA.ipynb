```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#reading excel file
car = pd.read_excel('/content/mtcars.xlsx')
car
```

```python
car.shape
```

```python
car.info()
```

```python
car.describe()
```

```python
# finding null values in the data
car.isna().sum()
```

```python
#replacing null values in wt column with central tendency(mean)
car['wt'].fillna(np.mean(car.wt), inplace= True)
```

```python
car.isna().sum()
```

```python
#replacing null values of gear column with mode
car['gear'].value_counts()
```

```python
car['gear'].fillna(3.0, inplace = True)
```

```python
car.isna().sum()
```

```python
#to check the impact of weight on mileage
plt.scatter(x = car.wt, y = car['mpg_high'])
plt.show()
```

```python
#To check impact of weight on displacement
plt.scatter(x = car.wt , y = car.disp)
plt.xlabel('Car Weight')
plt.ylabel('Displacement')
plt.show()
```

```python
sns.FacetGrid(car, hue = 'model')\
.map(plt.scatter, 'wt', 'mpg_city')
plt.show()
```

```python
#Finding number of cylinder in cars
gears = car['am'].value_counts()
```

```python
leb = ['Manual', 'Automatic']
plt.pie(gears.values, labels= leb)
plt.legend()
plt.show()
```

```python
plt.bar(leb, gears.values)
plt.show()
```
